## Cleaning 'Unknown' Values

### unknown values happens in 3 columns: job, education and contact

#### filling unknown in columns job and education

```python
def impute_null(cols):
    job = cols['job']
    
    if job != 'unknown':
        if job == 'unknown':
            return 'Null'
        return job
    
def impute_null2(cols):
    edu = cols['education']
    
    if edu != 'unknown':
        if edu == 'unknown':
            return 'Null'
        return edu

# def impute_null3(cols):
#     con = cols['contact']
    
#     if con != 'unknown':
#         if con == 'unknown':
#             return 'Null'
#         return con
```

```python
df['job'] = df.apply(impute_null, axis=1)
df['education'] = df.apply(impute_null2, axis=1)
# df['contact'] = df.apply(impute_null3, axis=1)
```

```python
jobList = []
for item in df['job'].unique():
    mean_bal = df[df['job']==item]['balance'].mean()
    mean_dur = df[df['job']==item]['duration'].mean()
    jobList.append([item, mean_bal, mean_dur])
dfjob = pd.DataFrame(columns=['id','mean_bal','mean_dur'],
                    data=jobList)
dfjob = dfjob.sort_values('mean_bal').reset_index().drop('index', axis=1)
dfjob = dfjob.drop(11)
dfjob
```

```python
dfjob['mean_bal'] = dfjob['mean_bal'].astype('int64')
dfjob['mean_dur'] = dfjob['mean_dur'].astype('int64')
dfjob
```

```python
def impute_student(cols):
    job = cols['job']
    mean_bal = dfjob['mean_bal']
    
    if (pd.isnull(job)):
        for i in range(0,len(dfjob)-1):
            if (cols['balance'] < mean_bal[0]):
                return dfjob['id'][0]
            if (mean_bal[i] < cols['balance'] < mean_bal[i+1]):
                return dfjob['id'][i]
            elif (cols['balance'] > mean_bal[len(dfjob)-1]):
                return dfjob['id'][len(dfjob)-1]
    else:
        return job
```

```python
df['job'] = df.apply(impute_student, axis=1)
```
