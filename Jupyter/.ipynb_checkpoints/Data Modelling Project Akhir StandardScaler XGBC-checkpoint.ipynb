# Label Encoder

```python
ec = LabelEncoder()
data['contact'] = ec.fit_transform(data['contact'])
data['default'] = ec.fit_transform(data['default'])
data['housing_loan'] = ec.fit_transform(data['housing_loan'])
data['personal_loan'] = ec.fit_transform(data['personal_loan'])
data['poutcome'] = ec.fit_transform(data['poutcome'])
data.head()
```

# One Hot Encoder

```python
job = pd.get_dummies(data['job'],drop_first=True)
marital = pd.get_dummies(data['marital'], drop_first=True)
education = pd.get_dummies(data['education'], drop_first=True)
month = pd.get_dummies(data['month'], drop_first=True)
```

```python
data.drop(['job','marital','education','month'],axis=1,inplace=True)
data_akhir= pd.concat([data,job,marital,education,month],axis=1)
pd.set_option('display.max_columns', None)
data_akhir.head()
```

# Modelling

```python
X1, X_test, y1, y_test = train_test_split(data_akhir, target, test_size=0.20,random_state=101)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X1, y1, test_size=0.25, random_state=101)
```

```python
ros = RandomOverSampler(random_state=101)
X_ros,y_ros = ros.fit_sample(X_train,y_train)
```

```python
X_ros = pd.DataFrame(X_ros, columns=X_train.columns)
y_ros= pd.Series(y_ros)
```

```python
# param_learning=(0.16,0.14,0.12)
# param_estimator=(250,300,350)
# param_grid = {'learning_rate' : param_learning, 'n_estimators' : param_estimator}

# xgb = XGBClassifier(random_state=101)
# dt = GridSearchCV(xgb, param_grid, scoring='roc_auc',cv=10, n_jobs=-1)
# dt.fit(X_ros, y_ros)
# print(dt.best_estimator_)
# print(dt.best_params_)
# print(dt.best_score_)
```

```python
def reporting(y_test, prediction):
    print(confusion_matrix(y_test, prediction))
    print(classification_report(y_test, prediction))
    print("ROC AUC : ", roc_auc_score(y_test, prediction))
    print("Accuracy : ", accuracy_score(y_test, prediction))
    print("Matthews Corr_coef : ",matthews_corrcoef(y_test, prediction))
```

```python
xgb1 = XGBClassifier(random_state=101,n_estimators=400,learning_rate=0.15, max_depth=4)
```

```python
xgb1.fit(X_ros,y_ros)
```

```python
predict = xgb1.predict(X_ros)
predProb = xgb1.predict_proba(X_ros)[:,1]
```

```python
reporting(y_ros,predict)
```

```python
predict2 = xgb1.predict(X_val)
predProb2 = xgb1.predict_proba(X_val)[:,1]
```

```python
reporting(y_val, predict2)
```

```python
predict3 = xgb1.predict(X_test)
predProb3 = xgb1.predict_proba(X_test)[:,1]
```

```python
reporting(y_test, predict3)
```

# Feature Importances

```python
feature_importance = xgb1.feature_importances_
feature_importance = 100.0 * (feature_importance/feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.figure(figsize=(20,16))
plt.subplot(1,2,2)
plt.barh(pos, feature_importance[sorted_idx], align='center')

plt.yticks(pos, data_akhir.columns.values[sorted_idx])
plt.xlabel("Relative Importance")
plt.title('Variable Importance')
plt.show()
```

```python
coef1 = pd.Series(xgb1.feature_importances_, data_akhir.columns).sort_values(ascending=False)
coef1.plot(kind='bar', title='feature importance')
```
