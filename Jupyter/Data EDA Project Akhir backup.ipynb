## Adding New Column for EDA

### For Range_Age

```python
def impute_range(cols):
    age = cols['age']
    
    if age < 20:
        return '18 - 20'
    elif 20 <= age <= 25:
        return '20 - 25'
    elif 26 <= age <= 30:
        return '26 - 30'
    elif 31 <= age <= 35:
        return '31 - 35'
    elif 36 <= age <= 40:
        return '36 - 40'
    elif 41 <= age <= 45:
        return '41 - 45'
    elif 46 <= age <= 50:
        return '46 - 50'
    elif 51 <= age <= 55:
        return '51 - 55'
    elif 56<= age <= 60:
        return '56 - 60'
    elif age > 60:
        return '60 - 99'
```

```python
df['range_age'] = df.apply(impute_range,axis=1)
```

```python
df.head()
```

### For Range_Duration

```python
def impute_duration(cols):
    dur = cols['duration']
    
    if 0 <= dur <= 350:
        return '0 - 350'
    elif 351 <= dur <= 700:
        return '350 - 700'
    elif 701 <= dur <= 1050:
        return '701 - 1050'
    elif 1051 <= dur <= 1400:
        return '1051 - 1400'
    elif dur > 1400:
        return '> 1400'
```

```python
df['range_dur'] = df.apply(impute_duration, axis=1)
```

```python
df.head()
```

### For Range_pdays

```python
def impute_days(cols):
    pdays = cols['pdays']
    
    if pdays < 0:
        return '0'
    elif 0 <= pdays <= 150:
        return '1 - 150'
    elif 151 <= pdays <= 300:
        return '151 - 300'
    elif 301 <= pdays <= 450:
        return '301 - 450'
    elif pdays > 451:
        return '> 451'
```

```python
df['range_pdays'] = df.apply(impute_days, axis=1)
df.head()
```
